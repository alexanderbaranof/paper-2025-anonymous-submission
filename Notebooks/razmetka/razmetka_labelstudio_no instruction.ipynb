```python
import pandas as pd
```

```python
df = pd.read_json('Data/processed_data/dataset.json', orient='index')
df = df[df['is_word_play']]
df = df[df['annotations'].apply(lambda x: len(x)>0)]
df
```

```python
df.annotations[3]
```

```python
def parse_annotations(i):

  text = ''

  for a in i:

    if type(a) == dict:

      headline_substring = a['headline_substring']
      reference_string = a['reference_string']
      wordplay_type = a['wordplay_type']

      if reference_string != '' and type(reference_string) != list:
        text = text + headline_substring + ' ----- ' + reference_string + ' ----- ' + wordplay_type + '\n'

      elif type(reference_string) == list:
        text = text + headline_substring + ' ----- ' +  ', '.join(reference_string) + ' ----- ' + wordplay_type + '\n'


    if type(a) == list:

      for x in a:
        headline_substring = x['headline_substring']
        reference_string = x['reference_string']
        wordplay_type = x['wordplay_type']

        if reference_string != '' and type(reference_string) != list:
          text = text + headline_substring + ' ----- ' + reference_string + ' ----- ' + wordplay_type + '\n'
        elif type(reference_string) == list:
          text = text + headline_substring + ' ----- ' +  ', '.join(reference_string) + ' ----- ' + wordplay_type + '\n'

  return text


df['labels'] = df['annotations'].apply(parse_annotations)
df
```

```python
df = df.query('labels != ""')
df
```

```python
for i in range(100):
  print(df.labels.iloc[i])
```

# Predictions extended promt

```python
models = ['gigachat_lite', 'gigachat_max', 'gpt_4o', 'mistral_nemo', 'yandex_gpt4']
predictions_extended = dict()
for model in models:
  preds = pd.read_json(f'Data/predictions/wordplay_interpretation_{model}_predictions.json', orient='index')
  if f'mistral_nemo_preds' in preds.columns:
    predictions_extended[model] = preds[f'mistral_nemo_preds'].apply(lambda i: i[0]['generated_text'][2]['content'])
  if f'gigachat_lite_preds' in preds.columns:
    predictions_extended[model] = preds[f'gigachat_lite_preds']
  elif f'gigachat_max_pred' in preds.columns:
    predictions_extended[model] = preds[f'gigachat_max_pred']
  elif 'gpt4o_explain' in preds.columns:
    predictions_extended[model] = preds[f'gpt4o_explain']
  elif 'yagpt_pred' in preds.columns:
    predictions_extended[model] = preds[f'yagpt_pred']
```

```python
df_to_razmetka = pd.DataFrame()
df_to_razmetka['headline'] = df['headline']
df_to_razmetka['lead'] = df['lead']
df_to_razmetka['labels'] = df['labels']
for k,v in predictions_extended.items():
 df_to_razmetka = df_to_razmetka.merge(v, left_index=True, right_index=True, how='left')

df_to_razmetka
```
